```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ==============================================================================
# CONFIGURAÇÃO INICIAL E BASE DE DADOS SIMULADA (Substitua pelos seus dados)
# Exemplo real: df = pd.read_csv('sua_base_de_dados.csv')
# ==============================================================================
np.random.seed(42)
dados = {
    'Departamento': np.random.choice(['Vendas', 'TI', 'Operações', 'Marketing'], 200),
    'Salario': np.random.normal(6500, 1500, 200),
    'Idade': np.random.randint(22, 60, 200),
    'Avaliacao_Desempenho': np.random.choice(['Excelente', 'Bom', 'Regular'], 200)
}
df = pd.DataFrame(dados)

# Configuração de estilo dos gráficos
sns.set_theme(style="whitegrid")

# ==============================================================================
# ATIVIDADE 01: ANÁLISES GRÁFICAS
# ==============================================================================

fig, axs = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Análise Exploratória de Dados - Visão Geral', fontsize=18, fontweight='bold')

# --- a) Gráfico de Setores (Pie Chart) ---
# Variável: Departamento
contagem_dept = df['Departamento'].value_counts()
cores_pie = sns.color_palette('pastel')[0:4]
axs[0, 0].pie(contagem_dept, labels=contagem_dept.index, autopct='%1.1f%%',
              colors=cores_pie, startangle=90, wedgeprops={'edgecolor': 'black'})
axs[0, 0].set_title('Distribuição de Funcionários por Departamento', fontsize=14)
axs[0, 0].legend(contagem_dept.index, title="Departamentos", loc="upper left", bbox_to_anchor=(1, 1))

# --- b) Gráfico de Barras (Bar Chart) ---
# Variável: Média de Salário por Avaliação de Desempenho
media_salario_aval = df.groupby('Avaliacao_Desempenho')['Salario'].mean().reset_index()
sns.barplot(data=media_salario_aval, x='Avaliacao_Desempenho', y='Salario',
            hue='Avaliacao_Desempenho', palette='viridis', ax=axs[0, 1], legend=True)
axs[0, 1].set_title('Média Salarial por Avaliação de Desempenho', fontsize=14)
axs[0, 1].set_xlabel('Avaliação de Desempenho', fontsize=12)
axs[0, 1].set_ylabel('Salário Médio (R$)', fontsize=12)

# --- c) Histograma ---
# Variável: Idade
sns.histplot(df['Idade'], bins=15, color='skyblue', kde=True, ax=axs[1, 0])
axs[1, 0].set_title('Distribuição da Idade dos Funcionários', fontsize=14)
axs[1, 0].set_xlabel('Idade (anos)', fontsize=12)
axs[1, 0].set_ylabel('Frequência (Qtd. Funcionários)', fontsize=12)

# --- d) Boxplot ---
# Variável: Salário
sns.boxplot(y=df['Salario'], color='lightgreen', ax=axs[1, 1], width=0.4)
axs[1, 1].set_title('Dispersão e Outliers dos Salários', fontsize=14)
axs[1, 1].set_ylabel('Salário (R$)', fontsize=12)
axs[1, 1].set_xlabel('Geral', fontsize=12)

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.savefig('graficos_analise.png') # Salva a imagem dos gráficos
plt.show()

# ==============================================================================
# ATIVIDADE 02: ANÁLISE UNIVARIADA (ESTATÍSTICA DESCRITIVA)
# ==============================================================================

def analise_univariada(coluna_nome, serie):
    print(f"\n{'-'*50}")
    print(f"ANÁLISE UNIVARIADA: {coluna_nome.upper()}")
    print(f"{'-'*50}")

    # a) Medidas de Tendência Central
    media = serie.mean()
    mediana = serie.median()
    moda = serie.mode()[0] # Pega a primeira moda caso haja mais de uma

    print("a) Medidas de Tendência Central:")
    print(f"   Média: {media:.2f}")
    print(f"   Mediana: {mediana:.2f}")
    print(f"   Moda: {moda:.2f}")

    # b) Medidas de Dispersão
    variancia = serie.var()
    desvio_padrao = serie.std()
    amplitude = serie.max() - serie.min()

    print("\nb) Medidas de Dispersão:")
    print(f"   Variância: {variancia:.2f}")
    print(f"   Desvio Padrão: {desvio_padrao:.2f}")
    print(f"   Amplitude: {amplitude:.2f}")

    # c) Medidas Separatrizes (Quartis)
    q1 = serie.quantile(0.25)
    q2 = serie.quantile(0.50) # Igual à mediana
    q3 = serie.quantile(0.75)

    print("\nc) Medidas Separatrizes (Quartis):")
    print(f"   Q1 (25%): {q1:.2f}")
    print(f"   Q2 (50%): {q2:.2f}")
    print(f"   Q3 (75%): {q3:.2f}")

# Realizando análise para duas variáveis diferentes
analise_univariada("Salário", df['Salario'])
analise_univariada("Idade", df['Idade'])
```
